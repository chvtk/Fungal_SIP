```python
%%R
library(phyloseq)
library(ggplot2)
library(foreach)
library(doParallel)
library(gridExtra)
library(vegan)
library(plyr)
library(RColorBrewer)
library(stats)
```

```python
%%R 
physeq = import_biom("data/otu_table_wtax.biom")
print(physeq)
sample.data = import_qiime_sample_data("data/Metadata_fungi.txt")
physeq = merge_phyloseq(physeq,sample.data)
physeq.md = sample_data(physeq)
physeq.fungi = prune_samples(physeq.md$Mock_Comm == "Y", physeq)
physeq.fungi.snorm = transform_sample_counts(physeq.fungi, function(x) x/sum(x))
physeq.fungi.thresh = filter_taxa(physeq.fungi, function(x) sum(x > 0) > (0.25 * length(x)), TRUE)
print(physeq.fungi.thresh)
```

```python
%%R

p = plot_bar(physeq.fungi.thresh, "Rank7", fill = "Rank2")
p
```
